```python
import os
import cv2
import numpy as np
from tqdm import tqdm

REBUILD_DATA = True

class Emotion():
    
    IMG_SIZE = 50
    
    NEUTRAL = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/neutral"
    CALM = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/calm"
    SAD = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/sad"
    HAPPY = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/happy"
    ANGRY = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/angry"
    FEARFUL = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/fearful"
    DISGUST = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/disgust"
    SURPRISED = "/Users/vineetkamat/Documents/College/Eight Semester/4701 AI Prac/voice/surprised"
    
    neutralcount = 0
    calmcount = 0
    sadcount = 0
    happycount = 0
    angrycount = 0
    fearfulcount = 0
    disgustcount = 0
    surprisedcount = 0
    
    
    LABELS = {NEUTRAL: 0, CALM: 1, SAD: 2, HAPPY: 3, ANGRY: 4, FEARFUL: 5, DISGUST: 6, SURPRISED: 7}
    training_data = []

    
    def make_training_data(self):
        for label in self.LABELS:
            print(label)
            for f in tqdm(os.listdir(label)):
                
                # lot of images have useless problems
                try:
                    path = os.path.join(label, f)
                    
                    '''
                    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
                    img = cv2.resize(img, (self.IMG_SIZE, self.IMG_SIZE))
                    self.training_data.append([np.array(img), np.eye(8)[self.LABELS[label]]])
                    '''
                    
                    scale, sr = librosa.load(path)
                    mel_spect = librosa.power_to_db(librosa.feature.melspectrogram(scale, sr=sr, n_fft=2048, hop_length=512, n_mels=10))
                    self.training_data.append([np.array(mel_spect), np.eye(8)[self.LABELS[label]]])
                    
                    
                    if label == self.NEUTRAL:
                        self.neutralcount += 1
                    elif label == self.CALM:
                        self.calmcount += 1
                    elif label == self.SAD:
                        self.sadcount += 1
                    elif label == self.HAPPY:
                        self.happycount += 1
                    elif label == self.ANGRY:
                        self.angrycount += 1
                    elif label == self.FEARFUL:
                        self.fearfulcount += 1
                    elif label == self.DISGUST:
                        self.disgustcount += 1
                    elif label == self.SURPRISED:
                        self.surprisedcount += 1
                        
                    
                except Exception as e:
                    print(e)
                    #pass
                    
        np.random.shuffle(self.training_data)
        np.save("voice_training_data.npy", self.training_data)
        print("Neutral: ", self.neutralcount)
        print("Calm: ", self.calmcount)
        print("Sad: ", self.sadcount)
        print("Happy: ", self.happycount)
        print("Angry: ", self.angrycount)
        print("Fearful: ", self.fearfulcount)
        print("Disgust: ", self.disgustcount)
        print("Surprised: ", self.surprisedcount)
        
if REBUILD_DATA:
    emotions = Emotion()
    emotions.make_training_data()
```
